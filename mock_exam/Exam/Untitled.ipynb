```python
image = pydicom.dcmread("ID_0062_AGE_0067_CONTRAST_0_CT.dcm")

images = []
for i, filename in enumerate(dcm_files):
    dcmimage = pydicom.dcmread(filename)
    age = int(dcmimage.PatientAge.replace("Y",""))
    filename_age = int(filename.split("_")[3])
    images.append((age, filename_age))
    if age != filename_age:
        print(f"Metadata does not match for {filename}")
    contrast = 0
    
    try: 
        if image.ContrastBolusAgent:
            contrast = 1
    except:
        AttributeError()

    print(f"ID_{i:04d}_AGE_{age:04d}_CONTRAST_{contrast}_CT.dcm")

image42 = pydicom.dcmread("ID_0042_AGE_0071_CONTRAST_1_CT.dcm")
image43 = pydicom.dcmread("ID_0043_AGE_0069_CONTRAST_1_CT.dcm")
```
